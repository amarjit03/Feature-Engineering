```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv("/content/SampleFile.csv")
```

# Maximum Scaling

```python
LotArea_MaxValue = np.max(df['LotArea'])
MSSubClass_MaxValue = np.max(df['MSSubClass'])
```

```python
df['LotArea_scaled'] = (df['LotArea']-LotArea_MaxValue)/LotArea_MaxValue
df['MSSubClass_scaled'] = (df['MSSubClass']-MSSubClass_MaxValue)/MSSubClass_MaxValue
```

```python
print(df)
```

```python
df_scaled = df[['LotArea_scaled','MSSubClass_scaled']].copy()
```

```python
print(df_scaled),print(df[['LotArea','MSSubClass']])
```

# Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df = df.drop(['LotArea_scaled','MSSubClass_scaled'],axis = 1)
```

```python
scaled_data = scaler.fit_transform(df)
scaled_df  = pd.DataFrame(scaled_data,columns = df.columns)
```

```python
scaled_df
```

# Normalization

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer()
```

```python
scaled_data = scaler.fit_transform(df)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data,columns = df.columns)
```

```python
print(scaled_df)
```

# Standardization
